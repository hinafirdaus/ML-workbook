```python
#creating random Normal distribution and establishing a relationship between PageSpeed and PurchaseAmount

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from pylab import *

pageSpeed= np.random.normal(3.0, 1.0, 1000)
PurchaseAmount = 100 - (pageSpeed + np.random.normal(0,0.1, 1000))*3
plt.xlabel("Page Speed")
plt.ylabel("Purchase Amount")
scatter(pageSpeed, PurchaseAmount)
```

```python
from scipy import stats
slope, intercept, r_val, p_val, std_err = stats.linregress(pageSpeed, PurchaseAmount)
```

```python
r_val**2
```

```python
def predict(x):
    return slope*x + intercept
fitline=predict(pageSpeed)
plt.scatter(pageSpeed, PurchaseAmount)
plt.plot(pageSpeed, fitline, c='r')
plt.show()
```
