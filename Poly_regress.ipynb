```python
#polyregreesion between the page speed and purchase amount using random normal distribution

%matplotlib inline
import matplotlib.pyplot as plt
from pylab import *
import numpy as np

np.random.seed(2)
pageSpeed=np.random.normal(3.0, 1.0,1000)
purchaseAmount = np.random.normal(50.0,10.0, 1000)/pageSpeed
scatter(pageSpeed, purchaseAmount)
```

```python
#for 4th poly degree let us see how the polyfit works

x=np.array(pageSpeed)
y=np.array(purchaseAmount)
p4=np.poly1d(np.polyfit(x,y,4))
```

```python
#lets create a regression to see the fitting

xp=np.linspace(0,7,100)
plt.xlabel("Page Speed 0-7 sec")
plt.ylabel("Purchase Amount")
plt.scatter(x,y)
plt.plot(xp, p4(xp), c='r')
plt.show()
```

```python
#r squared error value

from sklearn.metrics import r2_score
r2=r2_score(y,p4(x))
r2
```
