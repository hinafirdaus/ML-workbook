```python
#using polynomial regression creating PageSpeed and PurchaseAmount normal distribution

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from pylab import *

np.random.seed(2)  #creating pseudo determinstic random number generator
PageSpeed = np.random.normal(3.0,1.0, 100)
PurchaseAmount = np.random.normal(50.0,30.0, 100)/PageSpeed
plt.xlabel("Page Speed")
plt.ylabel("Purchase Amount")
scatter(PageSpeed, PurchaseAmount)
```

Let us divide the data into training and testing set with 80-20 ratio respectively. This ratio certainily will avoid overfitting

```python
trainX = PageSpeed[:80]
testX = PageSpeed[80:]

trainY=PurchaseAmount[:80]
testY=PurchaseAmount[80:]
```

```python
scatter(trainX,trainY)
```

```python
scatter(testX,testY)
```

```python
x=np.array(trainX)
y=np.array(trainY)
p8=np.poly1d(np.polyfit(x,y,8))
```

```python
xp=np.linspace(0,7,100)
axes=plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0,200])
plt.scatter(x,y)
plt.plot(xp,p8(xp), c='r')
plt.show()
```

```python
#calculating the r sqaured value for error

from sklearn.metrics import r2_score
r2= r2_score(testY,p8(testX))
print(r2)
```

```python
r2=r2_score(np.array(trainY),p8(np.array(trainX)))
print(r2)
```
